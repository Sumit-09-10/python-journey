### Question 17
**Topic:** isinstance()   
Write a function check_type(value) that returns the string 'number' if the value is an int or float (but not bool), and 'other' otherwise. Test it on 5, 5.5, True, and 'text'.

````python
def check_type(value):
    if isinstance(value, bool):
        return 'other'
    elif isinstance(value, (int, float)):
        return 'number'
    else:
        return 'other'

for v in [5, 5.5, True, 'text']:
    print(check_type(v))
````

### Question 70
**Topic:** Nested if  
**Difficulty:** Medium

Write code that classifies n = 15 as 'positive even', 'positive odd', 'negative even', 'negative odd', or 'zero'.

**Example Input:**  
```
n = 15
```

**Expected Output:**  
```
positive odd
```



````python
n = 15
if n == 0:
    print('zero')
elif n > 0:
    if n % 2 == 0:
        print('positive even')
    else:
        print('positive odd')
else:
    if n % 2 == 0:
        print('negative even')
    else:
        print('negative odd')
````

### Question 72

Write a function absolute_value(n) that returns the absolute value of n using only a conditional expression (no abs()). Test it on -7 and 4.

**Example Input:**  
```
-7, 4
```

**Expected Output:**  
```
7
4
```

````python
def absolute_value(n):
    return n if n >= 0 else -n

print(absolute_value(-7))
print(absolute_value(4))
````

### Question 95
**Topic:** Recursion  
Write a recursive function factorial(n) that returns n! (factorial). Test with n=6.

````python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(6))
````

### Question 99
**Topic:** nonlocal keyword  
Write a closure with an outer function make_counter() that contains an inner function count(). Use nonlocal to let count() modify the outer variable n. Return count and call it 3 times, printing n each time.

````python
def make_counter():
    n = 0
    def count():
        nonlocal n
        n += 1
        print(n)
    return count

counter = make_counter()
counter()
counter()
counter()
````

### Question 102
**Topic:** Interview: Fibonacci (recursion vs iteration)    
Write both a recursive and an iterative function to return the nth Fibonacci number. Test with n=10.

````python
# Recursive (clean but slow for large n)
def fib_recursive(n):
    if n <= 1:
        return n
    return fib_recursive(n - 1) + fib_recursive(n - 2)

# Iterative (efficient)
def fib_iterative(n):
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

print(fib_recursive(10))
print(fib_iterative(10))
````
